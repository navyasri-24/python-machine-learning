```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
col_names=['company','job','degree','salary_more_than_100k']
data=pd.read_csv("/content/salaries.csv",header=None,names=col_names)
data
```

```python
feature_cols=['company','job','degree']
x=data[feature_cols]
y=data['salary_more_than_100k']
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
data['company']=label_encoder.fit_transform(data['company'])
data['job']=label_encoder.fit_transform(data['job'])
data['degree']=label_encoder.fit_transform(data['degree'])
print(data.head())
```

```python
feature_cols=['company','job','degree']
x=data[feature_cols]
y=data['salary_more_than_100k']
#x=data.values[1:,:3]
#y=data.values[1:,3]
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

**CLUSTERING**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
plt.rcParams['figure.figsize']= (16,9)
x,y=make_blobs(n_samples=800,n_features=3,centers=4)
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss_list=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  wcss_list.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss_list)
plt.title("the elbow method")
plt.xlabel("number of clusters")
plt.ylabel("wcss_list")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
column_names=['user_id','item_id','rating','timestamp']
df=pd.read_csv('/content/u.data',sep='\t',names=column_names)
df.head()
```

```python
movie_titles=pd.read_csv("/content/Movie_Id_Titles")
movie_titles.head()
```

```python
df=pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating',y='num of ratings',data=ratings,alpha=0.5)
```

```python
moviemat=df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
starwars_user_ratings=moviemat['Star Wars (1977)']
liarliar_user_ratings=moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars=moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar=moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars=corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_liarliar=pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar=corr_liarliar.join(ratings['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
